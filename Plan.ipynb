# DataSet 
我目前使用的是MNIST数据集作为GroundTruth，使用Blender渲染出漫反射图像作为网络输入，图像数目10000张，分辨率800，800
因为MNIST数据集是黑白的，相对渲染出来的图像也应是黑白，所以对数据集灰度处理，然后分辨率缩小到126，126
